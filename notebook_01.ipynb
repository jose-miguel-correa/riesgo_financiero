SEGUIMOS PREPROCESANDO DATOS

```python
datos['emp_length'].unique()
```

```python
datos['emp_length_num'] = datos['emp_length'].replace('< 1 year', '0')
datos['emp_length_num'] = datos['emp_length_num'].str.extract('(\d+)', expand=False)
datos['emp_length_num'] = datos['emp_length_num'] .replace(np.nan, str(0))
datos['emp_length_num'] = pd.to_numeric(datos['emp_length_num'])
```

```python
datos['emp_length_num'].unique()
```

PREPROCESAMIENTO DE VARIABLE FECHA issue_d (fecha emisión de crédito)

```python
datos['issue_d']
```

```python
datos['fecha_credito'] = pd.to_datetime(datos['issue_d'], format=None)
datos['fecha_credito']
```

```python
pd.Timestamp.now() - datos['fecha_credito']
```

```python
datos['meses_credito'] = (pd.Timestamp.now() - datos['fecha_credito']).dt.total_seconds() / (30 * 24 * 60 * 60)
datos['meses_credito']
```
